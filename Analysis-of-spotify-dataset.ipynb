```python
#top 5 artist based on number of songs produced
top_five_artists=df.groupby("artist").count().sort_values(by="song name",ascending=False)['song name'][:5]
top_five_artists
```

```python
#pie chart representation of genre
df_genre = df['top genre'].value_counts()/len(df)
label=df_genre.index.tolist()
size=df_genre.values.tolist()
fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(size, labels=label, autopct='%1.1f%%', shadow=False, textprops={'fontsize': 8})
ax1.axis('equal')
plt.show()
```

```python
#bar graph representation of number of songs produced from year 2013 to 2017
year_song=df.groupby('year').count().sort_values(by='song name',ascending=True)['song name']
fig=year_song.plot.barh()
fig.set_xlabel('Frequency of songs released')
fig.set_ylabel('Year')
plt.show()
```

```python
top_loudness=df[['dB','song name']].sort_values(by='dB',ascending=False)[:20]
top_loudness.set_index('song name')
```

```python
#graphical representation of top 20 loudest tracks with their track names 
x=top_loudness['song name'].tolist()
y=top_loudness['dB'].tolist()
ax=sns.barplot(x=x,y=y)
ax.set_ylabel('Loudness')
ax.set_xticklabels(x,rotation=90)

plt.show()
```
